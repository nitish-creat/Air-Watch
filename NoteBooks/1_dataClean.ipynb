# Data Cleaning and Preparation

Dataset: Indian Air Quality Monitoring Data
- Dimensions: 425,279 records across 8 columns
- Key Metrics: AQI values, pollutants, air quality status by state and area

```python
import pandas as pd
import numpy as np
import os

data = pd.read_csv("../Data/MainDataset.csv")
print("Dataset loaded successfully!")
print(f"Shape: {data.shape}")
print(f"\nFirst few rows:")
print(data.head())
```

## Dataset Overview

```python
# print("Data Info:")
# print(data.info())
# print("\n" + "="*50)
print("Missing Values:")
print(data.isnull().sum())
print("\n" + "="*50)
print("Statistical Summary:")
print(data[['aqi_value', 'number_of_monitoring_stations']].describe())
```

## Data Cleaning

```python
data["date"] = pd.to_datetime(data["date"], format="%m/%d/%Y")
print("✓ Date converted to datetime format")

data = data.drop_duplicates()
print(f"✓ Duplicates removed")

data["year"] = data["date"].dt.year
data["month"] = data["date"].dt.month
data["day"] = data["date"].dt.day
data["day_of_week"] = data["date"].dt.day_name()
print("✓ Temporal features extracted (year, month, day, day_of_week)")

data["state"] = data["state"].str.strip().str.lower()
data["area"] = data["area"].str.strip().str.lower()
print("✓ State and area cleaned (lowercase, whitespace removed)")

aqi_status_mapping = {
    "Good": 1,
    "Satisfactory": 2,
    "Moderate": 3,
    "Poor": 4,
    "Very Poor": 5,
    "Severe": 6
}
data["quality_level"] = data["air_quality_status"].map(aqi_status_mapping)
print("✓ Air quality status encoded to numeric levels")

print("\nCleaned data info:")
print(data.info())
```

## Save Cleaned Data

```python
print("="*50)
print("Final Dataset Summary")
print("="*50)
print(f"Total records: {len(data)}")
print(f"Shape: {data.shape}")
# print(f"\nColumns: {list(data.columns)}")
# print(f"\nData types:\n{data.dtypes}")
# print(f"\nFirst 3 rows:")
# print(data.head(3))
print(f"\nMissing values:")
print(data.isnull().sum())
output_dir = "../Data/processed"
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, "clean_data.csv")
data.to_csv(output_path, index=False)
print(f"\n✓ Cleaned dataset saved to: {output_path}")
```
